```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
'''Read the ‘flavors of cocoa.csv’ file as a dataframe ‘df cocoa’ and answer questions 6-9. The description of features/variables is given below:'''
df=pd.read_csv('/content/flavors_of_cocoa.csv')
df
```

```python
''''Which of the following variable have null values?
 ID
 Company
 Review Date
 Rating'''
print("Columns in the DataFrame:")
print(df.columns)


null_summary = df.isnull().sum()
print("\nNull values in each column:")
print(null_summary)


null_columns = null_summary[null_summary > 0]
print("\nColumns with null values:")
print(null_columns)
```

```python
'''Which of the following countries have maximum locations of cocoa manufacturing companies?
 U.K.
 U.S.A.
 Canada
 France'''
location_counts = df['Company Location'].value_counts()

# Display the count of locations for each country
print(location_counts)
```

```python
'''After checking the data summary, which feature requires a data conversion considering the data values held?
 Rating
 Review date
 Company
 Bean origin'''
# Convert 'Review Date' to datetime format
df['Review Date'] = pd.to_datetime(df['Review Date'])
print(df['Review Date'])
```

```python
'''What is the maximum rating of chocolates?
 1.00
 5.00
 3.18
 4.00'''
# Assuming df_cocoa is already loaded
max_rating = df['Rating'].max()
print(f"The maximum rating of chocolates is: {max_rating}")
```
